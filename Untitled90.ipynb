```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cleaned_Stock_db.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
df.drop(columns= ['high', 'low', 'adjclose'], inplace = True)
df.head()
```

```python
df['MA_20'] = df['close'].rolling(window=20).mean()
df['MA_50'] = df['close'].rolling(window=50).mean()
df['MA_200'] = df['close'].rolling(window=200).mean()

# print("Moving Average 20 days", df['MA_20'])
# print("Moving Average 50 days", df['MA_50'])
# print("Moving Average 200 days", df['MA_200'])
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df['Date'], df['close'], label='Close Price', color='blue', linewidth=1.5)
plt.plot(df['Date'], df['MA_20'], label='20-day Moving Average', color='orange')
plt.plot(df['Date'], df['MA_50'], label='50-day Moving Average', color='green')
plt.plot(df['Date'], df['MA_200'], label='200-day Moving Average', color='red')

plt.title('Infosys Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['MA_20'] = df['close'].rolling(window=20).mean()
df['MA_50'] = df['close'].rolling(window=50).mean()

# Signal Column
df['Signal'] = 'HOLD'  # Default value

# Generate Buy/Sell signals
for i in range(1, len(df)):
    if df['MA_20'].iloc[i-1] < df['MA_50'].iloc[i-1] and df['MA_20'].iloc[i] > df['MA_50'].iloc[i]:
        df.loc[df.index[i], 'Signal'] = 'BUY'
    elif df['MA_20'].iloc[i-1] > df['MA_50'].iloc[i-1] and df['MA_20'].iloc[i] < df['MA_50'].iloc[i]:
        df.loc[df.index[i], 'Signal'] = 'SELL'

# Filter only signals
signal_df = df[df['Signal'].isin(['BUY', 'SELL'])]

# Plotting
plt.figure(figsize=(15, 8))
plt.plot(df['Date'], df['close'], label='Close Price', color='blue', alpha=0.6)
plt.plot(df['Date'], df['MA_20'], label='20-day MA', color='orange')
plt.plot(df['Date'], df['MA_50'], label='50-day MA', color='green')

# Plot Buy and Sell signals
plt.scatter(signal_df[signal_df['Signal'] == 'BUY']['Date'],
            signal_df[signal_df['Signal'] == 'BUY']['close'],
            marker='^', color='green', label='BUY Signal', s=100)

plt.scatter(signal_df[signal_df['Signal'] == 'SELL']['Date'],
            signal_df[signal_df['Signal'] == 'SELL']['close'],
            marker='v', color='red', label='SELL Signal', s=100)

plt.title('Infosys Buy/Sell Signals using MA Crossovers')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Create % return after 10 days from Buy/Sell point
df['Future_Close_10'] = df['close'].shift(-10)
df['Return_10D'] = ((df['Future_Close_10'] - df['close']) / df['close']) * 100

# Get only Buy/Sell rows
signal_df = df[df['Signal'].isin(['BUY', 'SELL'])].copy()

# Total signals
total_buy = (signal_df['Signal'] == 'BUY').sum()
total_sell = (signal_df['Signal'] == 'SELL').sum()

# Average return after 10 days of Buy/Sell
avg_buy_return = signal_df[signal_df['Signal'] == 'BUY']['Return_10D'].mean()
avg_sell_return = signal_df[signal_df['Signal'] == 'SELL']['Return_10D'].mean()

# Print summary
print(f" Total BUY Signals: {total_buy}")
print(f" Total SELL Signals: {total_sell}")
print(f" Avg Return 10 Days After BUY: {avg_buy_return:.2f}%")
print(f" Avg Return 10 Days After SELL: {avg_sell_return:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Your existing code for trend columns
df['Market_Trend'] = df['close'] > df['MA_200']
df['Trend_Label'] = df['Market_Trend'].apply(lambda x: 'Bullish' if x else 'Bearish')

# Make sure your 'Month' column is categorical with proper calendar order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# Group by month and trend, count the days
monthly_counts = df.groupby(['Month', 'Trend_Label']).size().unstack(fill_value=0)

# Plotting month-wise bullish vs bearish counts as line plot
plt.figure(figsize=(9, 4))

# Plot each trend as a separate line
plt.plot(monthly_counts.index, monthly_counts['Bullish'], marker='o', color='green', label='Bullish')
plt.plot(monthly_counts.index, monthly_counts['Bearish'], marker='o', color='red', label='Bearish')

plt.title('Month-wise Bullish vs Bearish Days in Infosys Stock', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Number of Days')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# Total bullish and bearish days overall
total_counts = df['Trend_Label'].value_counts()
print("Overall Market Trend Counts:")
print(total_counts)

# Monthly bullish and bearish counts
print("\nMonthly Counts:")
print(monthly_counts)

# Month with max bullish days
max_bullish_month = monthly_counts['Bullish'].idxmax()
print(f"\nMonth with highest bullish days: {max_bullish_month}")

# Month with max bearish days
max_bearish_month = monthly_counts['Bearish'].idxmax()
print(f"Month with highest bearish days: {max_bearish_month}")

# Bullish to bearish ratio per month
monthly_counts['Bullish_to_Bearish_Ratio'] = monthly_counts['Bullish'] / monthly_counts['Bearish'].replace(0, 1)
print("\nBullish to Bearish Ratio by Month:")
print(monthly_counts['Bullish_to_Bearish_Ratio'])
```

```python
from scipy.stats import chi2_contingency

# contingency table: rows=months, columns=Bullish and Bearish counts
contingency_table = monthly_counts[['Bullish', 'Bearish']]

# perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("Result: Significant association between Month and Market Trend (reject null hypothesis)")
else:
    print("Result: No significant association (fail to reject null hypothesis)")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Example for January and July bullish days
jan_bullish = monthly_counts.loc['Jan', 'Bullish']
jan_total = monthly_counts.loc['Jan'].sum()

jul_bullish = monthly_counts.loc['Jul', 'Bullish']
jul_total = monthly_counts.loc['Jul'].sum()

counts = [jan_bullish, jul_bullish]
nobs = [jan_total, jul_total]

stat, pval = proportions_ztest(counts, nobs)
print(f"Z-statistic: {stat:.4f}")
print(f"P-value: {pval:.4f}")

if pval < 0.05:
    print("Significant difference in bullish proportions between January and July")
else:
    print("No significant difference in bullish proportions between January and July")
```

```python
from statsmodels.stats.proportion import proportion_confint

# Overall bullish proportion
total_bullish = df['Trend_Label'].value_counts()['Bullish']
total_days = len(df)

lower, upper = proportion_confint(total_bullish, total_days, alpha=0.05, method='wilson')
print(f"95% CI for overall bullish proportion: ({lower:.3f}, {upper:.3f})")
```
